# Kullanıcı tabanlı önerme

```python
x = df.groupby('User-ID').count()['Book-Rating']>200
```

```python
filtered_rating = df[df['User-ID'].isin(x[x].index)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(y[y].index)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

## Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6]
    
    data = []
    
    for i in similar_books:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        
        data.append(item)
    return data
```

```python
recommend("1984")
```

```python
recommend("Year of Wonders")
```
